```python
data_dir = './dataset_without_bmi/torso'
```

```python
# split images in data_dir into train, validation and test sets
import os
import random
import shutil

# create directories for train, validation and test sets
train_dir = os.path.join(data_dir, 'train')
validation_dir = os.path.join(data_dir, 'validation')
test_dir = os.path.join(data_dir, 'test')

# list all images in data_dir
images = os.listdir(data_dir)
random.shuffle(images)
images
```

```python

# split images into train, validation and test sets
train_size = int(0.6 * len(images))
validation_size = int(0.2 * len(images))
test_size = len(images) - train_size - validation_size

train_images = images[:train_size]
validation_images = images[train_size:train_size+validation_size]
test_images = images[train_size+validation_size:]
```

```python
# copy images to train directory
os.makedirs(train_dir, exist_ok=True)
for image in train_images:
    shutil.copy(os.path.join(data_dir, image), os.path.join(train_dir, image))
```

```python
# copy images to validation directory
os.makedirs(validation_dir, exist_ok=True)
for image in validation_images:
    shutil.copy(os.path.join(data_dir, image), os.path.join(validation_dir, image))
```

```python
# copy images to test directory
os.makedirs(test_dir, exist_ok=True)
for image in test_images:
    shutil.copy(os.path.join(data_dir, image), os.path.join(test_dir, image))
```

```python
# list image in each directory
train_images = os.listdir(train_dir)
validation_images = os.listdir(validation_dir)
test_images = os.listdir(test_dir)

print('Number of images in train set:', len(train_images))
print('Number of images in validation set:', len(validation_images))
print('Number of images in test set:', len(test_images))
```

```python
# remove images in data_dir
for image in images:
    os.remove(os.path.join(data_dir, image))
```
