```python
import os
import cv2
```

```python
# Define the source and destination directory
src_directory = "outputs"
des_directory = "dataset"
```

```python
# Function to extract text from file path
def extract_filename(file_path):
  parts = file_path.split('/')[3].split('.')
  return parts[0]

def extract_bodypart(filename):
  parts = filename.split('_')
  
  if len(parts) >= 2:
    return parts[1]
  else:
    return parts[0]
  
def extract_bmi(file_path):
  parts = file_path.split('/')
  bmi = parts[2].split('_')
  return bmi[1]
  
def extract_folder(file_path):
  parts = file_path.split('/')
  return parts[1]
      
def save_image(image, folder, filename, bmi, des_directory):
  path = f'./{des_directory}/{folder}'
  path = os.path.join(path, extract_bodypart(filename))
  
  if not os.path.exists(path):
    os.makedirs(path)
    
  cnt = len(os.listdir(path))

  cv2.imwrite(filename=f'{path}/{cnt + 1}_{bmi}.jpg', img=image)
```

```python
# Iterate through all files in the directory and its subdirectories
for root, dirs, files in os.walk(src_directory):
  for file in files:
    if file == '.DS_Store':
      continue
    
    file_path = os.path.join(root, file)

    image = cv2.imread(file_path)
    
    filename = extract_filename(file_path)
    folder = extract_folder(file_path)
    bmi = extract_bmi(file_path)
    print(filename, folder)
    save_image(image=image, folder=folder, filename=filename, bmi=bmi, des_directory=des_directory)
```
