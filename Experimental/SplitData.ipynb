```python
# split folders into train, val and test folders
import os
import random
import shutil

# set the random seed
random.seed(0)

# define the folder to split
folder = './dataset_without_bmi/torso'

# define the ratio of train, val and test
train_ratio = 0.7
val_ratio = 0.15
test_ratio = 0.15

# define the folder to save the split dataset
train_folder = './dataset_without_bmi/torso/train'
val_folder = './dataset_without_bmi/torso/val'
test_folder = './dataset_without_bmi/torso/test'

# create the folders
if not os.path.exists(train_folder):
    os.makedirs(train_folder)
if not os.path.exists(val_folder):
    os.makedirs(val_folder)
if not os.path.exists(test_folder):
    os.makedirs(test_folder)

# get the list of files
file_list = os.listdir(folder)
file_list = [f for f in file_list if f.endswith('.jpg')]
num_files = len(file_list)

# shuffle the file list
random.shuffle(file_list)

# split the file list
train_list = file_list[:int(train_ratio*num_files)]
val_list = file_list[int(train_ratio*num_files):int((train_ratio+val_ratio)*num_files)]
test_list = file_list[int((train_ratio+val_ratio)*num_files):]

# copy the files to the corresponding folders
for f in train_list:
    shutil.copy(os.path.join(folder, f), os.path.join(train_folder, f))
for f in val_list:
    shutil.copy(os.path.join(folder, f), os.path.join(val_folder, f))
for f in test_list:
    shutil.copy(os.path.join(folder, f), os.path.join(test_folder, f))

print('Splitting is done!')
print('Train folder: {}'.format(train_folder))
print('Val folder: {}'.format(val_folder))
print('Test folder: {}'.format(test_folder))
print('Number of files in train folder: {}'.format(len(train_list)))
print('Number of files in val folder: {}'.format(len(val_list)))
print('Number of files in test folder: {}'.format(len(test_list)))
```
